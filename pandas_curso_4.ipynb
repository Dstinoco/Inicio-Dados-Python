```python
import pandas as pd
import numpy as np
```

```python
hospedagem_df = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json')
hospedagem_df.head()
```

```python
imoveis_dispobiveis_df = pd.read_json("https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json")
imoveis_dispobiveis_df.head()
```

```python
hospedagem_df = pd.json_normalize(hospedagem_df['info_moveis'])
hospedagem_df.head()
```

```python
colunas = list(hospedagem_df.columns)
colunas
```

```python
hospedagem_df = hospedagem_df.explode(colunas[3:])
hospedagem_df.head()
```

```python
hospedagem_df.reset_index(inplace=True, drop=True)
hospedagem_df.head()
```

```python
hospedagem_df['max_hospedes'] = hospedagem_df['max_hospedes'].astype('Int64')
hospedagem_df.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
hospedagem_df[col_numericas] = hospedagem_df[col_numericas].astype('Int64')
hospedagem_df['avaliacao_geral'] = hospedagem_df['avaliacao_geral'].astype('Float64')
```

```python
hospedagem_df.info()
```

```python
hospedagem_df['preco'] = hospedagem_df['preco'].apply(lambda x : x.replace('$', '').replace(",", '').strip())
hospedagem_df['preco']
```

```python
hospedagem_df['preco'] = hospedagem_df['preco'].astype('float64')
```

```python
hospedagem_df[['taxa_deposito', 'taxa_limpeza']] = hospedagem_df[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(",", '').strip())
```

```python
hospedagem_df[['taxa_deposito', 'taxa_limpeza']] = hospedagem_df[['taxa_deposito', 'taxa_limpeza']].astype('float64')
```

```python
hospedagem_df['descricao_local'] = hospedagem_df['descricao_local'].str.lower()
```

```python
hospedagem_df['descricao_local'] = hospedagem_df['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
hospedagem_df['descricao_local'] = hospedagem_df['descricao_local'].str.split()
hospedagem_df['descricao_local']
```

```python
hospedagem_df['comodidades'] = hospedagem_df['comodidades'].str.replace('\{|}|\"', '', regex=True)
hospedagem_df['comodidades'] = hospedagem_df['comodidades'].str.split(',')
```

```python
hospedagem_df['descricao_vizinhanca'] = hospedagem_df['descricao_vizinhanca'].str.split(',')
```
