```python
import pandas as pd
```

```python
vendas_clientes = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json')
dados_locacao_imoveis = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
```

```python
vendas_clientes = pd.json_normalize(vendas_clientes['dados_vendas'])
vendas_clientes.head()
```

```python
vendas_clientes = vendas_clientes.explode(['Cliente','Valor da compra'])
vendas_clientes.head()
```

```python
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].apply(lambda x: x.upper().replace('@', '').strip())
vendas_clientes.head()
```

```python
vendas_clientes['Valor da compra'] = vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(",", '').strip())
vendas_clientes.head()
```

```python
vendas_clientes.info()
```

```python
vendas_clientes['Data de venda'] = pd.to_datetime(vendas_clientes['Data de venda'])
vendas_clientes['Valor da compra'] = vendas_clientes['Valor da compra'].astype('float64')
```

```python
vendas_clientes.reset_index(inplace=True, drop=True)
```

```python
vendas_clientes.head()
```

```python
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis = pd.json_normalize(dados_locacao_imoveis['dados_locacao'])
dados_locacao_imoveis.head()
```

```python
colunas = list(dados_locacao_imoveis.columns)
```

```python
dados_locacao_imoveis = dados_locacao_imoveis.explode(['datas_combinadas_pagamento', 'datas_de_pagamento', 'valor_aluguel'])
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis['datas_de_pagamento'] = pd.to_datetime(dados_locacao_imoveis[ 'datas_de_pagamento'], format='%d/%m/%Y')
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '').replace('reais', '').strip())
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].astype('float64')
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis.reset_index(inplace=True, drop=True)
dados_locacao_imoveis.head()
```
