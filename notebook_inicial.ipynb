# Conhecendo a base de dados

## Importando os dados


```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep=";")
```

```python
dados.head(10)
```

```python
dados.info()
```

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.groupby("Tipo").mean(numeric_only = True).round(2)
```

```python
dados.groupby("Tipo")[['Valor']].mean().round(2).sort_values("Valor")
```

```python
df_preco_tipo = dados.groupby("Tipo")[['Valor']].mean().round(2).sort_values("Valor")

df_preco_tipo.plot(kind='barh', figsize=(14,10), color='red', title="Média Preço Imóveis")
```

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
df_residencial = dados.query("@imoveis_comerciais not in Tipo")
df_residencial.Tipo.unique()
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df_preco_tipo = df_residencial.groupby("Tipo")[['Valor']].mean().round(2).sort_values("Valor")

df_preco_tipo.plot(kind='barh', figsize=(14,10), color='red', title="Média Preço Imóveis")
```

```python
df_residencial[['Tipo']].value_counts(normalize=True).to_frame().sort_values("Tipo")
```

```python
df_percentual_tipo = df_residencial[['Tipo']].value_counts(normalize=True).to_frame().sort_values("Tipo")

df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color='green', 
                        xlabel="Tipos", ylabel='Perncentual', title="Média Preço Imóveis")
```

### **Selecionando apenas os imóveis do tipo apartamento**

```python
df_residencial.query("Tipo == 'Apartamento'")
```

```python
df = df_residencial.query("Tipo == 'Apartamento'")
df.head()
```

# Tratando e filtrando os dados

## Lidando com dados nulos

```python
df.isnull().sum().to_frame()
```

```python
df.fillna(0)
```

```python
df = df.fillna(0)
df.isnull().sum()
```

## Removendo registros

```python
df.query("Valor == 0 | Condominio == 0")
```

```python
removendo_df =df.query("Valor == 0 | Condominio == 0").index
```

```python
df.drop(removendo_df, axis=0, inplace=True)
```

```python
df.drop('Tipo', axis=1, inplace=True)
df.head()
```

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
selecao = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df["Area"] > 70)
df[selecao]
```

```python
df2 = df[selecao]
```
